# Python Performance Benchmark
This ipynb can be used to measure time required to run multiple code cells. This can be used to compare diffrent tools, environments and also to perform server stress test.

This test requires about 600MB of free RAM and 1 to 10 minutes (or even more on slow computers)

```python
# Loading required dependences
import time
```

## First test: Loading libraries

```python
start = time.time()
import numpy
time_numpy = time.time()-start

start = time.time()
import matplotlib.pyplot as plt
time_matplotlib = time.time()-start

start = time.time()
import pandas
time_pandas = time.time()-start

start = time.time()
import scipy
time_scipy = time.time()-start

start = time.time()
import random
time_random = time.time()-start
```

## Second test: Writing, Reading, Plotting 100 MB dataset

```python
# Write 100 MB random
file = open("data100mb.csv","w")
random.seed(1100)
start = time.time()
for i in range (3780000):
    print(i,random.random(),file=file, sep=",")
time_write_rand = time.time()-start
file.close()
```

```python
# Read 100 MB
start = time.time()
file = open("data100mb.csv","r")
times = []
values = []
for line in file:
    line = line.split(",")
    times.append(line[0])
    values.append(line[1])
file.close()
time_read = time.time()-start
```

```python
# Plot
start = time.time()
plt.scatter(times[1:3780],values[1:3780])
plt.show()
```

```python
time_plot = time.time()-start
```

```python
print("Library load time")
print("Pandas:",time_pandas)
print("NumPy:",time_numpy)
print("SciPy:",time_scipy)
print("Random:",time_random)
print("Matplotlib.pyplot as plt:",time_matplotlib)
print()
print("File manipulation time")
print("Write 100MB RANDOM:",time_write_rand)
print("Read 100MB:",time_read)
print("Plot:",time_plot)
print()
print("Total score:", (time_pandas + time_numpy + time_scipy+ time_random+time_matplotlib)*20 +time_write_rand + time_read + time_plot)
print("Smaller score = better")
```
